```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split


def load_data(train_data_path, weather_data_path):
    """
    :param data_path: path to the data
    :return: the data set after the pre-processing
    """
    train_data = pd.read_csv(train_data_path, low_memory=False)
    weather_data = pd.read_csv(weather_data_path, low_memory=False)
    return train_data, weather_data


def fill_none_values(df):
    # Filling None values
    df['max_temp_f_x'] = np.where(df['max_temp_f_x'] == 'None', 0.0, df['max_temp_f_x'])
    df['min_temp_f_x'] = np.where(df['min_temp_f_x'] == 'None', 0.0, df['min_temp_f_x'])
    df['precip_in_x'] = np.where(df['precip_in_x'] == 'None', 0.0, df['precip_in_x'])
    df['snow_in_x'] = np.where(df['snow_in_x'] == 'None', 0.0, df['snow_in_x'])
    df['snowd_in_x'] = np.where(df['snowd_in_x'] == 'None', 0.0, df['snowd_in_x'])
    df['max_wind_speed_kts_x'] = np.where(df['max_wind_speed_kts_x'] == 'None', 0.0,
                                          df['max_wind_speed_kts_x'])
    df['max_wind_gust_kts_x'] = np.where(df['max_wind_gust_kts_x'] == 'None', 0.0,
                                         df['max_wind_gust_kts_x'])
    df['avg_wind_speed_kts_x'] = np.where(df['avg_wind_speed_kts_x'] == 'None', 0.0,
                                          df['avg_wind_speed_kts_x'])
    df['avg_wind_drct_x'] = np.where(df['avg_wind_drct_x'] == 'None', 0.0,
                                     df['avg_wind_drct_x'])

    df['max_temp_f_y'] = np.where(df['max_temp_f_y'] == 'None', 0.0, df['max_temp_f_y'])
    df['min_temp_f_y'] = np.where(df['min_temp_f_y'] == 'None', 0.0, df['min_temp_f_y'])
    df['precip_in_y'] = np.where(df['precip_in_y'] == 'None', 0.0, df['precip_in_y'])
    df['avg_wind_speed_kts_y'] = np.where(df['avg_wind_speed_kts_y'] == 'None', 0.0,
                                          df['avg_wind_speed_kts_y'])
    df['avg_wind_drct_y'] = np.where(df['avg_wind_drct_y'] == 'None', 0.0,
                                     df['avg_wind_drct_y'])
    df['snow_in_y'] = np.where(df['snow_in_y'] == 'None', 0.0, df['snow_in_y'])
    df['snowd_in_y'] = np.where(df['snowd_in_y'] == 'None', 0.0, df['snowd_in_y'])
    df['max_wind_speed_kts_y'] = np.where(df['max_wind_speed_kts_y'] == 'None', 0.0,
                                          df['max_wind_speed_kts_y'])
    df['max_wind_gust_kts_y'] = np.where(df['max_wind_gust_kts_y'] == 'None', 0.0,
                                         df['max_wind_gust_kts_y'])

    df['CRSElapsedTime'].fillna(0.0, inplace=True)
    df['CRSElapsedTime'].fillna(0.0, inplace=True)
    df['Distance'].fillna(0.0, inplace=True)

    df['max_temp_f_x'].fillna(0.0, inplace=True)
    df['min_temp_f_x'].fillna(0.0, inplace=True)
    df['precip_in_x'].fillna(0.0, inplace=True)
    df['avg_wind_speed_kts_x'].fillna(0.0, inplace=True)
    df['avg_wind_drct_x'].fillna(0.0, inplace=True)
    df['snow_in_x'].fillna(0.0, inplace=True)
    df['snowd_in_x'].fillna(0.0, inplace=True)
    df['max_wind_speed_kts_x'].fillna(0.0, inplace=True)
    df['max_wind_gust_kts_x'].fillna(0.0, inplace=True)

    df['min_temp_f_y'].fillna(0.0, inplace=True)
    df['max_temp_f_y'].fillna(0.0, inplace=True)
    df['precip_in_y'].fillna(0.0, inplace=True)
    df['snow_in_y'].fillna(0.0, inplace=True)
    df['snowd_in_y'].fillna(0.0, inplace=True)
    df['max_wind_speed_kts_y'].fillna(0.0, inplace=True)
    df['max_wind_gust_kts_y'].fillna(0.0, inplace=True)
    df['avg_wind_speed_kts_y'].fillna(0.0, inplace=True)
    df['avg_wind_drct_y'].fillna(0.0, inplace=True)


def preprocess(train_data_path, weather_data_path):
    t_data, w_data = load_data(train_data_path, weather_data_path)

    # Convert the delay factor labels to categories, where -1 is no delay.
    t_data['DelayFactor'] = t_data['DelayFactor'].astype('category')
    t_data['DelayFactor'] = t_data['DelayFactor'].cat.codes

    # Drop samples where there is no delay but delay time is greater than 0
    t_data = t_data.drop(t_data[(t_data['DelayFactor'] == -1) & (t_data['ArrDelay'] > 0)].index)

    # Set Dates correctly
    t_data['FlightDate'] = pd.to_datetime(t_data['FlightDate'], errors='coerce')
    t_data.set_index(['FlightDate', 'Origin'])
    w_data['day'] = pd.to_datetime(w_data['day'], format='%d-%m-%y', errors='coerce')
    w_data.set_index(['day', 'station'])

    # Drop irrelevant features
    w_data = w_data.drop(columns=['max_dewpoint_f', 'min_dewpoint_f', 'min_rh', 'avg_rh', 'max_rh',
                                  'climo_high_f', 'climo_low_f', 'climo_precip_in', 'min_feel',
                                  'avg_feel', 'max_feel'])
    t_data = t_data.drop(columns=['Flight_Number_Reporting_Airline', 'OriginCityName',
                                  'OriginState', 'DestCityName', 'DestState', 'CRSArrTime'])

    # Rename the join columns
    w_data = w_data.rename(
        columns={'day': 'FlightDate', 'station': 'Origin'})
    # First merge to get the Origin weather data
    df_merge_once = pd.merge(t_data, w_data, on=['FlightDate', 'Origin'])
    # Rename to get the Dest
    w_data = w_data.rename(
        columns={'Origin': 'Dest'})
    # Second merge to get the Dest weather data
    df = pd.merge(df_merge_once, w_data, on=['FlightDate', 'Dest'])

    # Prepare categorical features
    df['FlightDate'] = pd.to_datetime(t_data['FlightDate'], errors='coerce')
    df['FlightYear'] = pd.DatetimeIndex(t_data['FlightDate']).year
    df['FlightMonth'] = pd.DatetimeIndex(t_data['FlightDate']).month

    CRSDepTime = pd.to_datetime(df['CRSDepTime'], format='%H%M', errors='coerce')
    CRSArrTime = pd.to_datetime(df['CRSArrTime'], format='%H%M', errors='coerce')
    df['CRSDepTime_hour'] = CRSDepTime.dt.hour
    df['CRSDepTime_min'] = CRSDepTime.dt.minute
    df['CRSArrTime_hour'] = CRSArrTime.dt.hour
    df['CRSArrTime_min'] = CRSArrTime.dt.minute

    # Create categorical features
    df = pd.get_dummies(df, columns=['DayOfWeek', 'Reporting_Airline', 'Origin', 'Dest',
                                     'FlightYear', 'FlightMonth', 'CRSDepTime_hour',
                                     'CRSDepTime_min', 'CRSArrTime_hour', 'CRSArrTime_min'])

    # Drop irrelevant features
    df = df.drop( columns=['Tail_Number', 'Flight_Number_Reporting_Airline', 'DestCityName',
                           'DestState', 'OriginCityName', 'OriginState', 'FlightDate', 'CRSDepTime',
                           'CRSArrTime'], axis=1)

    fill_none_values(df)

    return df

preprocess("train_data.csv", "all_weather_data.csv")
```
